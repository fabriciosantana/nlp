## III.1 Reconhecimento de Entidades Nomeadas

**Definio: Reconhecimento de Entidades Nomeadas (NER)**

O objetivo do NER  transformar dados textuais no estruturados em informaes estruturadas e facilmente interpretveis por sistemas computacionais.

**Importncia:**

O NER  uma ferramenta fundamental em diversas aplicaes de PLN devido  sua capacidade de extrair informaes essenciais de grandes volumes de texto de forma automatizada. Algumas das principais reas de aplicao incluem:

##III.2. Anlise de Sentimentos

**Conceito de Anlise de Sentimentos**

A anlise de sentimentos, tambm conhecida como minerao de opinio,  um campo do processamento de linguagem natural que envolve a identificao e extrao de informaes subjetivas de textos. O objetivo principal  determinar o estado emocional ou a opinio expressa em uma frase ou documento, classificando-o como positivo, negativo ou neutro. Esse processo  realizado atravs de algoritmos que analisam palavras, frases e contextos dentro de um texto para avaliar os sentimentos subjacentes.

**Importncia da Anlise de Sentimentos**

A anlise de sentimentos tem se tornado cada vez mais relevante em diversas reas, como negcios, marketing, poltica e pesquisa social. Aqui esto alguns motivos que destacam sua importncia:

1.   **Tomada de Decises Informadas:** Empresas utilizam a anlise de sentimentos para entender melhor a opinio de seus clientes sobre produtos e servios, ajudando na tomada de decises estratgicas.
2.   **Monitoramento de Marca:** A reputao de uma marca pode ser monitorada em tempo real, permitindo que as empresas respondam rapidamente a crises e gerenciem a percepo pblica.
3.   **Aprimoramento de Produtos:** Feedbacks negativos identificados atravs da anlise de sentimentos podem indicar reas de melhoria, ajudando as empresas a aprimorarem seus produtos ou servios.
4.   **Anlise de Competio:** As opinies dos consumidores sobre concorrentes podem ser analisadas para identificar pontos fortes e fracos relativos.
5.   **Engajamento com o Cliente:** A anlise de sentimentos ajuda a personalizar a comunicao com os clientes, ajustando mensagens para serem mais empticas e relevantes.


**Classificao na Anlise de Sentimentos**

A anlise de sentimentos geralmente classifica o texto em trs categorias principais:


1.   Positivo: Indica uma opinio favorvel ou emoes positivas, como alegria, entusiasmo e satisfao.
2.   Negativo: Indica uma opinio desfavorvel ou emoes negativas, como tristeza, raiva e frustrao.
3.   Neutro: Indica uma opinio neutra ou a ausncia de emoo significativa, podendo incluir fatos ou declaraes objetivas.

Alguns sistemas de anlise de sentimentos avanados tambm identificam sentimentos mais especficos, como medo, surpresa, confiana, entre outros.

**Interpretao dos Resultados**

A interpretao dos resultados da anlise de sentimentos envolve a compreenso das mtricas fornecidas pelos algoritmos. As principais mtricas incluem:

1.   **Polaridade:** Uma pontuao que varia de -1 a 1, onde -1 indica uma opinio muito negativa, 1 indica uma opinio muito positiva e 0 indica neutralidade.
2.   **Subjectividade:** Uma pontuao que varia de 0 a 1, onde 0 representa fatos objetivos e 1 representa opinies subjetivas.
3.   **Pontuaes Positivas, Negativas e Neutras:** Propores que indicam a presena de sentimentos positivos, negativos e neutros no texto analisado.

Por exemplo, uma anlise de sentimentos de uma avaliao de produto pode resultar em uma polaridade de 0.8 (muito positiva), com uma subjectividade de 0.9 (altamente subjetiva), indicando uma opinio pessoal muito favorvel ao produto.

**Aplicaes Prticas**

1.   **Redes Sociais:** Monitoramento de menes de marca e anlise de feedback dos clientes.
2.   **Servio ao Cliente:** Anlise de interaes para melhorar o atendimento ao cliente e identificar problemas recorrentes.
3.   **Poltica:** Anlise de discursos, debates e opinies pblicas para compreender o sentimento popular em relao a polticas e candidatos.
4.   **Pesquisa de Mercado:** Compreenso de tendncias e preferncias do consumidor atravs da anlise de opinies em reviews, blogs e fruns.

A anlise de sentimentos  uma ferramenta que transforma dados textuais em insights acionveis, permitindo que empresas e organizaes compreendam melhor as opinies e emoes de seus pblicos. Sua capacidade de fornecer informaes em tempo real e detalhadas sobre a percepo de marca, satisfao do cliente e tendncias do mercado torna-se essencial para a tomada de decises estratgicas e a melhoria contnua dos servios e produtos oferecidos.

```python
# Verificar e instalar googletrans e NLTK
try:
    from googletrans import Translator
    import nltk
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    googletrans_installed = True
except ImportError:
    googletrans_installed = False

if not googletrans_installed:
    !pip install googletrans==4.0.0-rc1
    !pip install nltk
    import nltk
    from googletrans import Translator
    from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Garantir que o vader_lexicon est instalado
nltk.download('vader_lexicon')

# Inicializar o analisador de sentimentos VADER
analyzer = SentimentIntensityAnalyzer()

# Inicializar o tradutor
translator = Translator()

# Textos de exemplo em portugus
texts = [
    "Eu estou muito feliz hoje, mas ontem foi um dia terrvel.",
    "O cliente reclamou do atraso na entrega das mercadorias."
]

# Funo para traduzir texto usando googletrans
def traduzir_texto(texto):
    try:
        traduzido = translator.translate(texto, dest='en')
        return traduzido.text
    except Exception as e:
        print("Erro na traduo:", e)
        return None

# Traduzir os textos para ingls usando googletrans e analisar os sentimentos
for text in texts:
    translated_text = traduzir_texto(text)
    if translated_text:
        # Analisar o sentimento do texto traduzido
        sentiment = analyzer.polarity_scores(translated_text)

        # Imprimir os resultados
        print(f"Texto original: {text}")
        print(f"Texto traduzido: {translated_text}")
        print(f"Polaridade: {sentiment['compound']}")
        print(f"Positivo: {sentiment['pos']}")
        print(f"Negativo: {sentiment['neg']}")
        print(f"Neutro: {sentiment['neu']}")
        print("\n")
    else:
        print(f"No foi possvel traduzir o texto: {text}")
```

##III.3 Desambiguao de Palavras com NLTK e Google Translate

A **desambiguao de palavras**  um processo crucial no campo do Processamento de Linguagem Natural que consiste em identificar qual sentido de uma palavra est sendo usado em um determinado contexto. Muitas palavras tm mltiplos significados, e a habilidade de determinar o sentido correto  essencial para a compreenso precisa de textos.

**Por exemplo,** a palavra "banco" pode se referir a uma instituio financeira ou a um assento. O contexto da frase determina qual sentido  apropriado.

**Importncia**

A desambiguao de palavras  importante por vrias razes:

1.   **Melhoria da Preciso em Aplicaes de PLN:** A desambiguao correta das palavras melhora a preciso de diversas aplicaes de PLN, como traduo automtica, busca em documentos e respostas automticas.
2.   **Anlise de Sentimentos:** Identificar o sentido correto das palavras ajuda a interpretar melhor os sentimentos expressos em textos.
3.   **Assistentes Virtuais:** Assistentes virtuais, como Siri e Alexa, dependem da desambiguao de palavras para compreender e responder adequadamente s solicitaes dos usurios.
4.   **Recuperao de Informao:** Sistemas de busca e recuperao de informao utilizam a desambiguao para fornecer resultados mais relevantes.

**Tarefa 3.1.** Explorar um exemplo prtico usando NLTK e Google Translate para desambiguar palavras em portugus. Este exemplo mostrar como determinar o sentido correto da palavra "banco" e "manga" em diferentes contextos. E para avaliar este cenrio:

```python
# Verificar e instalar googletrans e NLTK

try:
    from googletrans import Translator
    import nltk
    from nltk.corpus import wordnet
    nltk_installed = True
except ImportError:
    nltk_installed = False

if not nltk_installed:
    !pip install googletrans==4.0.0-rc1
    !pip install nltk
    import nltk
    from googletrans import Translator
    from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

# Inicializar o tradutor
translator = Translator()

# Funo para desambiguao de palavras
def get_sense(word, context):
    synsets = wordnet.synsets(word, lang='por')
    if not synsets:
        return None
    best_sense = synsets[0]
    max_overlap = 0

    for synset in synsets:
        definition = synset.definition().split()
        overlap = len(set(definition).intersection(set(context)))
        if overlap > max_overlap:
            max_overlap = overlap
            best_sense = synset

    return best_sense

# Funo para traduzir texto usando googletrans
def traduzir_texto(texto, dest_lang='pt'):
    try:
        traduzido = translator.translate(texto, dest=dest_lang)
        return traduzido.text
    except Exception as e:
        print("Erro na traduo:", e)
        return texto

# Exemplos de textos em portugus
examples = [
    ("Ele vestiu uma camisa com uma longa manga.", "manga"),
    ("Ele foi ao banco para depositar dinheiro.", "banco")
]

# Processar cada exemplo
for text, word in examples:
    context = text.split()
    sense = get_sense(word, context)
    if sense:
        translated_definition = traduzir_texto(sense.definition(), 'pt')
        print(f"Texto: {text}")
        print(f"Palavra: {word}")
        print(f"Sentido encontrado: {sense.name()}")
        print(f"Definio em ingls: {sense.definition()}")
        print(f"Definio traduzida: {translated_definition}")
        print("\n")
    else:
        print(f"No foi possvel encontrar um sentido para a palavra: {word}")
        print("\n")
```

## III.4 Resoluo de Correferncia

**Definio**

Correferncia  um conceito no campo do Processamento de Linguagem Natural que se refere  identificao de todas as expresses em um texto que se referem  mesma entidade. Em outras palavras,  a tarefa de determinar quais palavras ou frases so co-referentes, ou seja, quais delas apontam para o mesmo objeto ou pessoa no mundo real.

**Por exemplo,** na frase "Maria foi ao mercado. Ela comprou frutas.", as palavras "Maria" e "Ela" so co-referentes, pois ambas se referem  mesma pessoa.

**Aplicao**

A resoluo de correferncia  uma tarefa essencial em vrias aplicaes de PLN, incluindo:

1.   **Anlise de Textos e Resumo Automtico:** Facilita a compreenso de quem ou o que est sendo referido ao longo de um documento, melhorando a qualidade dos resumos gerados automaticamente.
2.   **Assistentes Virtuais e Chatbots:** Melhora a capacidade de assistentes virtuais como Siri, Alexa e chatbots de entender e responder a perguntas sequenciais, mantendo a coerncia e a relevncia no contexto da conversa.
3.   **Traduo Automtica:** Ajuda na traduo de textos mantendo a consistncia das referncias ao longo do texto traduzido, garantindo que pronomes e outros termos referenciem corretamente as entidades mencionadas.
4.   **Anlise de Sentimentos e Extrao de Informaes:** Permite uma anlise mais precisa dos sentimentos e a extrao de informaes especficas sobre entidades, como pessoas, empresas e produtos, em textos como avaliaes de clientes e notcias.

**Importncia**

A resoluo de correferncia  importante por vrias razes:

1.   **Melhoria da Coerncia Textual:** Identificar corretamente as referncias ajuda a manter a coerncia ao longo do texto, essencial para a compreenso humana e para o processamento automatizado.
2.   **Preciso na Extrao de Informaes:** Permite extrair informaes de forma mais precisa, associando corretamente aes, atributos e emoes s entidades apropriadas.
3.   **Aprimoramento da Interao Humano-Computador:** Melhora a interao com assistentes virtuais e chatbots, tornando-os mais eficientes e naturais ao manter o contexto e as referncias corretas ao longo da conversa.
4.   **Qualidade na Traduo de Textos:** Garante que as tradues mantenham a consistncia das referncias, evitando ambiguidades e erros de interpretao.

```python
# Verificar e instalar as bibliotecas necessrias

try:
    import accelerate
    import transformers
    import pandas as pd
    import torch
    import sklearn
    libraries_installed = True
except ImportError:
    libraries_installed = False

if not libraries_installed:
    !pip install accelerate -U
    !pip install transformers[torch] -U
    !pip install pandas
    !pip install torch
    !pip install scikit-learn

# Verificar novamente aps a instalao
try:
    import accelerate
    import transformers
    import pandas as pd
    import torch
    import sklearn
    print("Todas as bibliotecas foram instaladas com sucesso.")
except ImportError as e:
    print(f"Ocorreu um erro ao instalar as bibliotecas: {e}")
```

```python
# Estas linhas importam as bibliotecas necessrias. As principais so:
# transformers: para carregar e treinar o modelo BERT.
# torch: uma biblioteca de aprendizado de mquina.
# pandas: para manipulao de dados.
# sklearn: para avaliao do modelo.

from transformers import BertTokenizer, BertForTokenClassification, Trainer, TrainingArguments
import torch
from torch.utils.data import Dataset
import pandas as pd
from sklearn.metrics import classification_report


# Definir dataset
class CorefDataset(Dataset):
    def __init__(self, sentences, labels, tokenizer, max_len):
        self.sentences = sentences
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.sentences)

    def __getitem__(self, item):
        sentence = str(self.sentences[item])
        labels = self.labels[item]

        encoding = self.tokenizer.encode_plus(
            sentence,
            max_length=self.max_len,
            add_special_tokens=True,
            truncation=True,
            padding='max_length',
            return_attention_mask=True,
            return_tensors='pt',
        )

        tokens = self.tokenizer.convert_ids_to_tokens(encoding['input_ids'][0])
        labels_vector = [0] * len(tokens)

        for i, token in enumerate(tokens):
            for label in labels:
                if label.lower() in token.lower():
                    labels_vector[i] = 1

        labels_vector = labels_vector + [0] * (self.max_len - len(labels_vector))

        return {
            'sentence_text': sentence,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(labels_vector[:self.max_len], dtype=torch.long)
        }

# Instanciar tokenizer
# Aqui, carregamos um tokenizador BERT pr-treinado especfico para a lngua
# portuguesa. O tokenizador divide as frases em pedaos menores que o modelo
# pode entender.
tokenizer = BertTokenizer.from_pretrained('neuralmind/bert-base-portuguese-cased')

# Criar dataset
# Usamos a classe CorefDataset para criar um dataset com nossas sentenas e
# rtulos. Aqui, df.sentence e df.labels so colunas de um dataframe df.
dataset = CorefDataset(
    sentences=df.sentence.to_numpy(),
    labels=df.labels.to_numpy(),
    tokenizer=tokenizer,
    max_len=32
)

# Dividir em treino e teste
# Dividimos o dataset em duas partes: uma para treinar o modelo (train_dataset)
# e outra para avaliar o modelo (val_dataset).
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

# Carregar modelo
# Carregamos um modelo BERT pr-treinado para a tarefa de classificao de tokens,
# que identifica diferentes tipos de palavras em uma sentena.
model = BertForTokenClassification.from_pretrained('neuralmind/bert-base-portuguese-cased', num_labels=2)

# Configurar treinamento
# Configuramos como o treinamento deve ser executado, incluindo o nmero de
# pocas (quantas vezes o modelo ver cada dado), tamanho do lote (quantas
# sentenas so processadas de uma vez), e outras configuraes de otimizao.
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=5,                 # nmero de pocas (hiperparmetro)
    per_device_train_batch_size=2,      # tamanho de lote (hiperparmetro)
    per_device_eval_batch_size=2,       # tamanho de lote (hiperparmetro)
    warmup_steps=10,
    weight_decay=0.01,
    learning_rate=5e-5,                 # taxa de aprendizado (hiperparmetro)
    logging_dir='./logs',
    save_strategy="epoch",
    report_to="none"  # add - Desativa integraes externas, incluindo wandb
)

# add - Funo de mtricas de avaliao
def compute_metrics(pred):
    labels = pred.label_ids.flatten()
    preds = pred.predictions.argmax(-1).flatten()
    return classification_report(labels, preds, output_dict=True)

# Instanciar Trainer
# Criamos um treinador (Trainer) que gerencia o treinamento do modelo,
# utilizando o dataset de treino e validao que criamos.
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)

# Treinar modelo
# Iniciamos o processo de treinamento, onde o modelo aprende a identificar e
# classificar os tokens nas sentenas.
trainer.train()

# Finalmente, salvamos o modelo treinado e o tokenizador para uso futuro.
model.save_pretrained('./results')
tokenizer.save_pretrained('./results')
```

**Resoluo de Correferncia Usando o Modelo Treinado**

Para tal, o programa a seguir realiza as etapas:

1.   **Importao de Ferramentas:** Carregar bibliotecas necessrias.
2.   **Carregar Modelo e Tokenizador:** Preparar o modelo para uso.
3.   **Funo de Coreferncia:** Analisar textos e identificar quais palavras se referem  mesma entidade.
4.   **Funo de Avaliao:** Verificar a preciso do modelo.
5.   **Teste e Avaliao:** Testar com um exemplo e avaliar com dados de validao.


```python
# Importao de Bibliotecas
#
# Transformers:
#
# pipeline, BertTokenizer, BertForTokenClassification: Estas ferramentas so
# usadas para carregar um modelo de machine learning (ML) chamado BERT,
# que pode entender e analisar textos.
#
# Scikit-Learn:
#
# classification_report: Ferramenta para avaliar a preciso do modelo de ML.

from transformers import pipeline, BertTokenizer, BertForTokenClassification
from sklearn.metrics import classification_report
import torch
import pandas as pd

# Carregar o modelo treinado e o tokenizer:
#tokenizer = BertTokenizer.from_pretrained('./results')
#model = BertForTokenClassification.from_pretrained('./results')

tokenizer = BertTokenizer.from_pretrained('neuralmind/bert-base-portuguese-cased')
model = BertForTokenClassification.from_pretrained('neuralmind/bert-base-portuguese-cased')

# Definir a funo de resoluo de correferncia:
# Funo resolve_coreference: Funo que analisa o texto e identifica as palavras
#                             que se referem  mesma entidade.
# Pipeline nlp: Uma linha de montagem que conecta o modelo e o tokenizador para
#               processar o texto.
# Tokenizao: Dividir o texto em partes menores (tokens).
# Predies (predictions): O modelo analisa os tokens e faz predies sobre
#                          quais palavras so co-referentes.
# Resultados (results): Lista de palavras do texto e seus rtulos preditos.
def resolve_coreference(text):
    nlp = pipeline("token-classification", model=model, tokenizer=tokenizer)
    tokens = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**tokens)
    predictions = torch.argmax(outputs.logits, dim=2)
    results = []
    for token, pred in zip(tokens.input_ids[0], predictions[0]):
        word = tokenizer.decode([token])
        results.append((word, pred.item()))
    return results

# Funo para avaliao:
# Funo evaluate_model: Avalia o quo bem o modelo est funcionando.
# y_true e y_pred: Listas que armazenam os rtulos verdadeiros e os rtulos
#                  preditos, respectivamente.
# Relatrio de Classificao: Ferramenta que mostra a preciso do modelo em
#                             identificar coreferncias.
def evaluate_model(dataset):
    y_true = []
    y_pred = []

    for data in dataset:
        sentence = data['sentence_text']
        true_labels = data['labels'].numpy()
        pred_labels = resolve_coreference(sentence)

        # Ajustar y_true e y_pred para a avaliao
        for true_label, (word, pred_label) in zip(true_labels, pred_labels):
            y_true.append(true_label)
            y_pred.append(pred_label)  # O segundo item  o rtulo predito

    print(classification_report(y_true, y_pred, target_names=['O', 'B-CORE']))

# Testar com um Texto de exemplo:
# Texto de Exemplo: Frase que ser analisada para resolver coreferncias.
# Resultado (result): Aplicao da funo de coreferncia no texto de exemplo.
# Exibir Resultado: Mostrar as palavras e seus rtulos em formato de tabela.
text = "Maria viu Joo no parque. Ela o cumprimentou calorosamente."

# Resolver correferncia
result = resolve_coreference(text)
df = pd.DataFrame(result, columns=["Word", "Label"])
print(df)

# Avaliar modelo:
# Avaliar o modelo usando um conjunto de dados de validao (val_dataset).
evaluate_model(val_dataset)
```

```python
import time
# --- Traduo Automtica com Google Translate API e Sada de Voz ---

# Verificar e instalar as bibliotecas necessrias
try:
    from googletrans import Translator
    from gtts import gTTS
    from IPython.display import Audio, display
    googletrans_installed = True
except ImportError:
    googletrans_installed = False

if not googletrans_installed:
    !pip install googletrans==4.0.0-rc1
    !pip install gtts
    from googletrans import Translator
    from gtts import gTTS
    from IPython.display import Audio, display

# Criar objeto de traduo
translator = Translator()

# Lista de frases em portugus
texts = [
    "O gato est no tapete.",
    "A casa  grande e bonita.",
    "Ele gosta de jogar futebol.",
    "Vamos ao cinema hoje  noite.",
    "Ela est estudando para o exame.",
    "O gato preto da Renata dorme neste momento."
]

# Traduzir textos para ingls e francs
translations_en = [translator.translate(text, src='pt', dest='en').text for text in texts]
translations_fr = [translator.translate(text, src='pt', dest='fr').text for text in texts]

# Funo para converter texto em udio e reproduzir
def text_to_speech(text, lang):
    tts = gTTS(text=text, lang=lang)
    tts.save("temp.mp3")
    display(Audio("temp.mp3", autoplay=False))

# Imprimir resultados e reproduzir udio
print("Tradues para o ingls:")
for text, translation in zip(texts, translations_en):
    print(f'Texto original: {text}')
    print(f'Traduo: {translation}')
    text_to_speech(translation, 'en')
    print()

print("Tradues para o francs:")
for text, translation in zip(texts, translations_fr):
    print(f'Texto original: {text}')
    print(f'Traduo: {translation}')
    text_to_speech(translation, 'fr')
    print()
```
